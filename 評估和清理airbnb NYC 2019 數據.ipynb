#### 3. æ•¸å€¼æ¬„ä½

- å‹æ…‹ç„¡éœ€è½‰æ›ï¼Œå¾ŒçºŒå°‡é‡å°ç•°å¸¸å€¼ï¼ˆå¦‚éé«˜çš„ `price` æˆ– `minimum_nights`ï¼‰é€²è¡Œè™•ç†ã€‚


#### 4.æ—¥æœŸæ¬„ä½

`last_review` ç›®å‰ç‚ºç‰©ä»¶å‹æ…‹ï¼ˆobject / stringï¼‰
é è¨ˆè½‰æ›ç‚º datetime å‹æ…‹ï¼Œæ–¹ä¾¿å¾ŒçºŒåšæ™‚é–“æ’åºã€åˆ†æœˆçµ±è¨ˆæˆ–è¨ˆç®—å¤©æ•¸


**ç¸½çµ**ï¼šä¸»è¦è½‰æ›é›†ä¸­åœ¨è­˜åˆ¥ç¢¼æ¬„ä½ (`id`ã€`host_id`) è½‰ç‚ºæ–‡å­—å‹æ…‹ï¼Œä¿ç•™æµ®é»å‹æ…‹æ¬„ä½ä»¥ä¾¿é‹ç®—ï¼ŒåŒæ™‚è™•ç†ç¼ºå¤±å€¼èˆ‡ç•°å¸¸å€¼ã€‚



```python
cleaned_data['id'] = cleaned_data['id'].astype(int).astype(str) #id
cleaned_data['host_id'] = cleaned_data['host_id'].astype(int).astype(str) #host_id

cleaned_data['last_review'] = pd.to_datetime(original_data['last_review']) #last_review è½‰æˆdatetime
```

```python
# 1ï¸âƒ£ ç¶“ç·¯åº¦ç¯„åœæª¢æŸ¥
cleaned_data = cleaned_data[
    (cleaned_data['latitude'].between(40.5, 41)) &
    (cleaned_data['longitude'].between(-74.25, -73.7))
]
```

```python
# 2ï¸âƒ£ price èˆ‡ minimum_nights ç•°å¸¸å€¼ç¯©é¸ï¼ˆIQR æ³•ï¼‰


# price
Q1_price = cleaned_data['price'].quantile(0.25)
Q3_price = cleaned_data['price'].quantile(0.75)
IQR_price = Q3_price - Q1_price
lower_price = Q1_price - 1.5 * IQR_price
upper_price = Q3_price + 1.5 * IQR_price

# minimum_nights
Q1_min = cleaned_data['minimum_nights'].quantile(0.25)
Q3_min = cleaned_data['minimum_nights'].quantile(0.75)
IQR_min = Q3_min - Q1_min
lower_min = Q1_min - 1.5 * IQR_min
upper_min = Q3_min + 1.5 * IQR_min

# ç¯©æ‰æ¥µç«¯å€¼
cleaned_data = cleaned_data[
    (cleaned_data['price'] >= lower_price) &
    (cleaned_data['price'] <= upper_price) &
    (cleaned_data['minimum_nights'] >= lower_min) &
    (cleaned_data['minimum_nights'] <= upper_min)
]

```
