# 項目：評估和清理英國電商公司銷售數據

## 分析目標

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

那麽針對此數據集，我們無需評估重覆數據。

接下來篩選出`UnitPrice`為負數的觀察值。

```python
original_df[original_df['UnitPrice'] < 0 ]
```

從輸出結果來看，`UnitPrice`為負數的觀察值都是壞賬調賬，不屬於實際商品交易數據，因此也在數據清理步驟中也將其刪除。

## 清理数据

根據前面評估部分得到的結論，我們需要進行的數據清理包括：
- 把`InvoiceDate`變量的數據類型轉換為為日期時間
- 把`CustomerID`變量的數據類型轉換為字符串
- 把`Description`變量缺失的觀察值刪除
- 把`Country`變量值`"USA"`替換為`"United States"`
- 把`Country`變量值`"UK"`、`"U.K."`替換為`"United Kingdom"`
- 把`Quantity`變量值為負數的觀察值刪除
- 把`UnitPrice`變量值為負數的觀察值刪除

為了區分開經過清理的數據和原始的數據，我們創建新的變量`cleaned_data`，讓它為`original_data`覆制出的副本。我們之後的清理步驟都將被運用在`cleaned_data`上。

```python
cleaned_data = original_df.copy()

cleaned_data.head()
```

把`InvoiceDate`變量的數據類型轉換為為日期時間類型：

```python
cleaned_data['InvoiceDate'] = pd.to_datetime(cleaned_data['InvoiceDate']) 
cleaned_data["InvoiceDate"]
```

把`CustomerID`變量的數據類型轉換為字符串：

## cleaned_data['CustomerID'] = cleaned_data['CustomerID'].astype(str)
cleaned_data['CustomerID']

把`CustomerID`变量值结尾的`".0"`删除：

```python
cleaned_data['CustomerID'] = cleaned_data['CustomerID'].str.slice(0,-2)
cleaned_data['CustomerID']
```

把`Description`變量缺失的觀察值刪除，並查看刪除後該列空缺值個數和：

```python
cleaned_data.dropna(subset=['Description'],inplace = True)
```

```python
cleaned_data['Description'].isnull().sum()
```

把`Country`變量值`"USA"`替換為`"United States"`，並檢查替換後`"USA"`變量值個數：

```python
cleaned_data['Country'].replace({"USA": "United States"},inplace=True)
```

把`Country`變量值`"UK"`、`"U.K."`替換為`"United Kingdom"`，並檢查替換後`"UK"`和`"U.K."`變量值個數：

```python
cleaned_data['Country'].replace(["U.K",'UK','U.K.'], "United Kingdom",inplace=True)
```

```python
print(len(cleaned_data[cleaned_data["Country"] == "US"]))  
```

```python
print(len(cleaned_data[cleaned_data["Country"] == "UK"]))  
print(len(cleaned_data[cleaned_data["Country"] == "U.K."]))
```

把`Quantity`變量值為負數的觀察值刪除，並檢查替換後`Quantity`變量值為負數的個數：

```python
cleaned_data = cleaned_data[cleaned_data['Quantity'] >= 0 ]
```

```python
len(cleaned_data[cleaned_data['Quantity'] < 0 ] )
```

把`UnitPrice`變量值為負數的觀察值刪除，並檢查替換後`UnitPrice`變量值為負數的個數：

```python
cleaned_data = cleaned_data[cleaned_data['UnitPrice'] >= 0 ]
```

```python
len(cleaned_data[cleaned_data['UnitPrice'] < 0 ] )
```
